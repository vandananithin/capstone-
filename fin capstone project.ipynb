# Statistical test

```python
cont_var=["iralcfy",'irmjfy', 'ircocfy', 'irhallucyfq', 'irinhalyfq','wrkdhrswk2','irherfy','irmethamyfq','cig30use']


#These are continous columns
```

```python
cat_var = []
for i in df.columns:
    if i not in cont_var:
        cat_var.append(i)
cat_var

#these are categorical columns
```

### CHISQUARE

```python
import scipy.stats as stats
```

```python
pval = []
for i in cat_var:
    obs = pd.crosstab(df1['addprev'],df1[i])
    stat,p,dof,expected =stats.chi2_contingency(obs)
    pval.append(p)
    print(p)
```

```python
chisquare_table =pd.DataFrame()
chisquare_table['cat_independent_features']=cat_var
chisquare_table['p-value']=pval
chisquare_table
```

```python
df1.shape
```

### t test independent

```python
p_val = []
for i in cont_var:
    s1 = df1[df1['addprev']==0.0][i]
    s2=df1[df1['addprev']==1.0][i]
    test_stat,pval=stats.ttest_ind(s1,s2)
    p_val.append(pval)
    print(i,":",pval)
```

```python
con_ttest=pd.DataFrame()
con_ttest['Features'] = cont_var
con_ttest['p-value'] = p_val
con_ttest
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,classification_report,roc_auc_score
```

```python
print(cat_var)
```

```python
#removing target varible from list
```

```python
cat_var = ['cigever', 'alcever', 'crkever', 'cocever', 'herever', 'methamevr', 'mjever', 'lsd', 'pcp', 'peyote', 'mesc', 'psilcy', 'ecstmolly', 'ketminesk', 'dmtamtfxy', 'salviadiv', 'hallucoth', 'inhalever', 'catag3', 'health', 'irwrkstat', 'ireduhighst2', 'newrace2', 'irsex', 'irpinc3', 'irki17_2', 'booked', 'probaton', 'drvinalco', 'drvinmarj', 'drvincocn', 'drvinhern', 'drvinhall', 'drvininhl', 'drvinmeth', 'mhsuipln', 'mhsuitry', 'irmarit', 'coutyp4', 'irhhsiz2', 'year']
```

```python

df_num  = df[cont_var]
df_cat = df[cat_var]

df_num_scale = SS.fit_transform(df_num)
df_num_scale_df = pd.DataFrame(df_num_scale,columns=df_num.columns)
df_cat_ecode = pd.get_dummies(df_cat,drop_first=True)
final_df = pd.concat([df_num_scale_df,df_cat_ecode],1)
```

```python
X = final_df
y =df['addprev']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10,test_size=0.3)
```

```python
lr=LogisticRegression(max_iter=10000)
lr_model=lr.fit(X_train,y_train)
y_pred_train=lr_model.predict(X_train)
y_pred_test=lr_model.predict(X_test)
y_pred_proba_train=lr_model.predict_proba(X_train)
y_pred_proba_test=lr_model.predict_proba(X_test)
print("Results on Train data: \n")
print("Accuracy on train data :",accuracy_score(y_train, y_pred_train))
print("\n Confusion matrix : \n",confusion_matrix(y_train,y_pred_train))
print("\n Classification report : \n",classification_report(y_train,y_pred_train))
print("ROC_AUC score :",roc_auc_score(y_train,y_pred_proba_train[:,1]))
print("---------------------------")
print("Results on Test data: \n")
print("Accuracy on test data :",accuracy_score(y_test, y_pred_test))
print("\n Confusion matrix : \n",confusion_matrix(y_test,y_pred_test))
print("\n Classification report : \n",classification_report(y_test,y_pred_test))
print("ROC_AUC score :",roc_auc_score(y_test,y_pred_proba_test[:,1]))
```

```python
cm_lr=confusion_matrix(y_test,y_pred_test)
# sensitivity
sen=cm_lr[1,1]/float(cm_lr[1,1]+cm_lr[1,0])
print(sen)
# specificity
spe=cm_lr[0,0]/float(cm_lr[0,0]+cm_lr[0,1])
print(spe)
```

```python
lr_df={'Model_Name':'Logistic Regression','Train_Accuracy':accuracy_score(y_train, y_pred_train)
      ,'Test_Accuracy':accuracy_score(y_test, y_pred_test),'Sensitivity':sen,'Specificity':spe}
score=pd.DataFrame(lr_df,index=[0])
score
```
